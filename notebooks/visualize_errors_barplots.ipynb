```python
total_sep_sep=np.load(testpath+'/total_sep_sep.npy', allow_pickle=True)
total_cls_sep=np.load(testpath+'/total_cls_sep.npy', allow_pickle=True)
total_ovlps_sep=np.load(testpath+'/total_ovlps_sep.npy', allow_pickle=True)

total_sep_bh=np.load(testpath+'/total_sep_bh.npy', allow_pickle=True)
total_cls_bh=np.load(testpath+'/total_cls_bh.npy', allow_pickle=True)
total_ovlps_bh=np.load(testpath+'/total_ovlps_bh.npy', allow_pickle=True)

sep_total = [i for i in range(4000) if distance[i] > 20.0]
cls_total = [i for i in  range(4000) if  6.0 < distance[i] <= 20.0]
ovlps_total = [i for i in  range(4000) if distance[i] <= 6.0 ]

acc_sepobj_bh = 1-(total_sep_bh/len(sep_total))
acc_sepobj_sep = 1-(total_sep_sep/len(sep_total))

acc_clsobj_bh = 1-(total_cls_bh/len(cls_total))
acc_clsobj_sep = 1-(total_cls_sep/len(cls_total))

acc_ovlpsobj_bh = 1-(total_ovlps_bh/len(ovlps_total))
acc_ovlpsobj_sep = 1-(total_ovlps_sep/len(ovlps_total))

#Compute mean acc 

mean_acc_sep_bh = [get_mean(data=acc_sepobj_bh[i]) for i in range(len(acc_sepobj_bh))]
mean_acc_sep_sep = [get_mean(data=acc_sepobj_sep[i]) for i in range(len(acc_sepobj_sep))]

mean_acc_cls_bh  = [get_mean(data=acc_clsobj_bh[i]) for i in range(len(acc_clsobj_bh))]
mean_acc_cls_sep = [get_mean(data=acc_clsobj_sep[i]) for i in range(len(acc_clsobj_sep))]

mean_acc_ovlps_bh = [get_mean(data=acc_ovlpsobj_bh[i]) for i in range(len(acc_ovlpsobj_bh))]
mean_acc_ovlps_sep = [get_mean(data=acc_ovlpsobj_sep[i]) for i in range(len(acc_ovlpsobj_sep))]

# Get stds

std_acc_sep_bh = [get_std(data=acc_sepobj_bh[i]) for i in range(len(acc_sepobj_bh))]
std_acc_sep_sep = [get_std(data=acc_sepobj_sep[i]) for i in range(len(acc_sepobj_sep))]

std_acc_cls_bh  = [get_std(data=acc_clsobj_bh[i]) for i in range(len(acc_clsobj_bh))]
std_acc_cls_sep = [get_std(data=acc_clsobj_sep[i]) for i in range(len(acc_clsobj_sep))]

std_acc_ovlps_bh = [get_std(data=acc_ovlpsobj_bh[i]) for i in range(len(acc_ovlpsobj_bh))]
std_acc_ovlps_sep = [get_std(data=acc_ovlpsobj_sep[i]) for i in range(len(acc_ovlpsobj_sep))]
```
